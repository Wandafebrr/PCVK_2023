````python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
````

````python
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
````

````python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255), thickness=15)
plt.imshow(black_img)
````

````python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0),thickness=8)
plt.imshow(black_img)
````

````python
cv.circle(black_img, center=(400,400), radius=50, color=(255,0,0),thickness=-1)
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
````

````python
from skimage.util.dtype import dtype_limits
black_img2=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
````

````python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
````

````python
pts=vertices.reshape((-1,1,2))  # nilai 2 untuk menunjukkan bahwa tiap titiik dibuat 3 cahnnel yg mewakili rgb
pts
````

````python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
````

1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib? <br>
**Jawab :**
Matplotlib melakukan pembacaan format simpan warna dengan urutan RGB. Urutan yang terbalik akan menghasilkan output yang keliru. Sehingga perlu dijalankannya perintah cv.cvtColor(img,cv.COLOR_BGR2RGB) untuk konversi dari format BGR (Blue-Green-Red) ke RGB (Red-Green-Blue) karena OpenCV menggunakan format 3 kanal warna tersebut dengan urutan BGR. Jika tidak dilakukan maka nilai intensitas pada channel Blue akan dianggap sebagai nilai intensitas pada channel Red, dan demikian seterusnya.

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?<br>
**Jawab :**
Perbedaan antara tipe data int16 dan int32 dalam konteks pembuatan black image terletak pada jumlah bit yang digunakan untuk merepresentasikan setiap elemen gambar. Ini akan berdampak pada rentang nilai yang dapat direpresentasikan oleh masing-masing tipe data dan penggunaan memori yang diperlukan. Jadi pengaruhnya akan terlihat terutama dalam hal ukuran memori yang digunakan dan rentang nilai yang dapat diterima. Penggunaan memori yang lebih rendah dengan int16 dapat menjadi pilihan yang baik untuk merepresentasikan gambar hitam sederhana atau jika memori menjadi pertimbangan penting.
